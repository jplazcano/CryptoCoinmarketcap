```python
import yfinance as yf  # Importar la biblioteca yfinance para descargar datos de Yahoo Finance
import pandas as pd   # Importar pandas para manipulación de datos
import numpy as np    # Importar numpy para cálculos numéricos
import matplotlib.pyplot as plt # Importar matplotlib para graficación
```

```python
tickers = ['AAPL', 'AMZN', 'TSLA', 'MSFT'] # Selección de tickers
start_date = '2018-01-01' # Fecha comienzo
end_date = '2023-07-31' # Fecha fin
```

```python
df = yf.download(tickers, start_date, end_date) # Descarga de datos
df # Imprimir los datos descargados
```

```python
df = df.filter(like='Adj Close')  # Seleccionar solo las columnas "Adj Close"
df # Imprimir df
```

```python
df.columns = df.columns.droplevel(0) # Eliminar el multi-índice
df # Imprimir df
```

```python
monthly_df = df.resample('BMS').first() # Crear variable Reorganizando los datos en un marco de tiempo mensual
monthly_df # Imprimir los datos mensuales
```

```python
# Calcular los retornos diarios y mensuales y guardarlos en variables
daily_returns = df.pct_change()
monthly_returns = monthly_df.pct_change()
```

```python
daily_returns # Imprimir los retornos diarios
```

```python
monthly_returns # Imprimir los retrornos mensuales
```

```python
# Eliminar los valores NaN
daily_returns = daily_returns.dropna()
monthly_returns = monthly_returns.dropna()
```

```python
daily_returns # Imprimir retornos diarios sin valores NaN
```

```python
monthly_returns # Imprimir retornos mensuales sin valores NaN
```

```python
covariances = {} # Diccionario para almacenar matrices de covarianza
daily_returns_id = daily_returns.index # Obtener los índices de las fechas de los retornos diarios para futuras comparaciones

# Iterar y calcular matrices de covarianza para cada mes
for i in monthly_returns.index:
  mask = (daily_returns_id.month == i.month) & (daily_returns_id.year == i.year) # Máscara para seleccionar el mes y año
  covariances[i] = daily_returns[mask].cov() # Calcular y almacenar matriz de covarianza
```

```python
covariances # Imprimir toda la variable covariances
```

```python
# Acceder a la matriz de covarianza correspondiente a la primera fecha en el diccionario de covarianzas
covariances[list(covariances.keys())[0]]
```

```python
# Diccionarios para almacenar retornos, volatilidad y pesos de cartera
portfolio_returns = {}
portfolio_volatility = {}
portfolio_weights = {}



for date in sorted(covariances.keys()): # Iterar a través de fechas
  cov = covariances[date]  # Obtener matriz de covarianza para la fecha
  for portfolio in range(5000): # Generar 5000 carteras aleatorias
    weights = np.random.random(4) # Generar pesos aleatorios
    weights /= np.sum(weights)  # Normalizar los pesos
    returns = np.dot(weights, monthly_returns.loc[date]) # Calcular retornos
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov, weights))) # Calcular volatilidad

    portfolio_returns.setdefault(date, []).append(returns) # Almacenar retornos
    portfolio_volatility.setdefault(date, []).append(volatility) # Almacenar volatilidad
    portfolio_weights.setdefault(date, []).append(weights) # Almacenar pesos
```

```python
# Graficar los portafolios
plt.figure(figsize=(10,8))
plt.scatter(x=portfolio_volatility[date], y=portfolio_returns[date], alpha=0.1)
plt.xlabel('Volatility')
plt.ylabel('Returns')
plt.show()
```

```python
# Función para mostrar los pesos de las carteras seleccionadas
def display_portfolio_weights(date, portfolio_idx):
  # Obtener pesos
  weights = portfolio_weights[date][portfolio_idx]
  # Imprimir retornos y volatilidad
  print(f"Portfolio {portfolio_idx} - Return: {portfolio_returns[date][portfolio_idx]:.4f},"
        f"Volatility: {portfolio_volatility[date][portfolio_idx]:.4f}")
  # Iterar a través de los pesos
  for ticker, weight in zip(tickers, weights):
    print(f"{ticker}: {weight:.4f}") # Imprimir ticker y peso
  print()
```

```python
# Obtener la última fecha de datos disponibles
date = sorted(covariances.keys())[-1]
date
```

```python
returns = portfolio_returns[date]
returns
```

```python
volatilities = portfolio_volatility[date]
volatilities
```

```python
# Graficar la frontera eficiente y carteras seleccionadas
plt.figure(figsize=(10,8))
plt.scatter(x=volatilities, y=returns, alpha=0.1) # Graficar todos los portafolios
plt.xlabel('Volatility')
plt.xlabel('Volatility')
plt.ylabel('Returns')

top_returns_idx = np.array(returns).argsort()[-1:][::-1] # Seleccionar portafolio con mayor retorno
low_volatilities_idx = np.array(volatilities).argsort()[:1] # Seleccionar portafolio con menor volatilidad

plt.scatter(x=np.array(volatilities)[top_returns_idx],
            y=np.array(returns)[top_returns_idx], color='r')  # Marcar portafolio con mayor retorno
plt.scatter(x=np.array(volatilities)[low_volatilities_idx],
            y=np.array(returns)[low_volatilities_idx], color='g') # Marcar portafolio con menor volatilidad
plt.show() # Mostrar el gráfico
```

```python
# Mostrar las carteras seleccionadas
print('Portfolio with top returns:')
for idx in top_returns_idx:
  print(f"Portfolio {idx}")
  display_portfolio_weights(date, idx)
print()
```

```python
# Mostrar las carteras seleccionadas
print('Portfolio with lowest volatility:')
for idx in low_volatilities_idx:
  print(f"Portfolio {idx}")
  display_portfolio_weights(date, idx)
print()
```

```python
# Calcular la envolvente convexa y mostrar portafolios eficientes aleatorios
from scipy.spatial import ConvexHull

volatilities = np.array(portfolio_volatility[date])
returns = np.array(portfolio_returns[date])

hull = ConvexHull(np.column_stack((volatilities, returns))) # Calcular envolvente convexa

efficient_frontier_idx = hull.vertices # Obtener índices de portafolios eficientes

random_efficient_portfolios = np.random.choice(efficient_frontier_idx, 10, replace=False) # Seleccionar portafolios eficientes aleatorios
```

```python
# Graficar los portafolios eficientes
plt.figure(figsize=(10,8))
plt.scatter(x=volatilities, y=returns, alpha=0.1)
plt.scatter(x=volatilities[random_efficient_portfolios], y=returns[random_efficient_portfolios],
            color='r', label="Random Efficient Portfolios")
plt.xlabel("Volatility")
plt.ylabel("Returns")
plt.show()
```
